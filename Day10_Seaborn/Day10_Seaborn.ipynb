Distribution plots in seaborn are distplot,joinplot,pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

If the values are floats or integers , then only we can find the corelation else we can't find , becuase categorical values are object type.Always corelation exists between -1 to +1.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

Joinplot allows to study the relationship between 2 numeric variables.The central chart display their correlation.It is usually
a scatterplot,a hexbin plot,a 2D histogram or a 2D density plot

```python
sns.jointplot(x = 'tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x = 'tip',y='total_bill',data=df,kind='reg')
```

Pair Plot is also know as scatterplot in which one variable in the same data row is matched with another variable's value
like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.Pair plots are ued when we have more than one independent variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

Dist plot helps us to check the distribution of the column feature.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

Categorical Plots are like boxplot,violinplot,countplot,bar plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y = 'sex',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot('sex','tip',data=df)
```

```python
sns.boxplot(x = 'day',y='tip',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
