```python
import matplotlib.pyplot as plt #This is an alias for using matplotlib

%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
#plt Scatter

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test1.png')
```

```python
#plt plot

y = x*x
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#Subplots - means creating multiple canvas in one single plot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x = np.arange(1,15)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
#Printing the sine wave
x = np.arange(0,4*np.pi,0.1)# the last value 0.1 is step value
y = np.sin(x)
plt.title("Sine wave form")

plt.plot(x,y)
plt.show()
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align = 'center')
plt.bar(x2,y2,color = 'g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
a = np.array([22,89,56,53,11,12,24,61,55,55,55])
plt.hist(a)
plt.title("histogram")
plt.show()
```

Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] # std indicates the standard devivation distribution

plt.boxplot(data,vert = False,patch_artist=False);
```

```python
#Pie Chart

labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,190]
colors = ['gold','yellow','red','blue']
explode = (0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
